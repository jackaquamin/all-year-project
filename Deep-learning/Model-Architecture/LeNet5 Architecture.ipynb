### LeNet5 Architecture


```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, AveragePooling2D, Activation, Dense, Flatten
```

```python
input_shape = (32,32,1)
num_class = 10

# Create model LeNet5
cnn_model = Sequential()

# Input Layer
# C1
cnn_model.add(Conv2D(6,(5,5), input_shape=input_shape))
# Activation
cnn_model.add(Activation('tanh'))

# P1
cnn_model.add(AveragePooling2D(pool_size=(2, 2), strides=2))
# Activation
cnn_model.add(Activation('tanh'))

# C2
cnn_model.add(Conv2D(16,(5,5)))
# Activation
cnn_model.add(Activation('tanh'))

# P2
cnn_model.add(AveragePooling2D(pool_size=(2, 2), strides=2))
# Activation
cnn_model.add(Activation('tanh'))

# C3
cnn_model.add(Conv2D(120,(5,5)))
# Activation
cnn_model.add(Activation('tanh'))

# Flatten
cnn_model.add(Flatten())

# Fully Connected Layer
cnn_model.add(Dense(84))
# Activation
cnn_model.add(Activation('tanh'))

# Output Layer
cnn_model.add(Dense(num_class))
# Activation
cnn_model.add(Activation('softmax')) # multi class classificaton (0-9)

cnn_model.summary()
```
